**Finding Missing Data**

```python
import numpy as np 
import pandas as pd 

# Search = .isna()
# drop = .dropna()
# Fill = .fillna()
```

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
# check if columns have any NULL value
df.isna().any()
```

**Removing Missing Data**

```python
df
```

```python
# .dropna()  work on row basis, it will drop row if it is containing any NULL val
df.dropna()
# it will not change in original data frame, just return a snapshot
# to do inplace make inplace=True
```

```python
df
```

```python
df.dropna(thresh=2) #thresh meaning there must be 2 non null values in each row, if any row doesn't have 2 non NULL it will remove that row
```

**Filling the missing Data**

```python
df
```

```python
df.fillna(0)
```

```python
# in column A it will fill 0, in B will fill 100 ...
vals = {"A":0 , "B":100, "C":200, "D":300}
df.fillna(value = vals)
```

```python
df
```

```python
df.fillna(df.mean())
```
