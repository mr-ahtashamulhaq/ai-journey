```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

### **Distribution Plots**
We use distribution plots to visualize the distribution of quantitative data / continuous data (data which cannot be divided into categories)

Let's discuss some plots that allow us to visualize the distribution of a data set.

These plots are:

1. Distplot/histplot
2. Jointplot
3. Pairplot
4. Rugplot

```python
df = sns.load_dataset('tips') # Seaborn has some builtin dataset so load one of them
df
```

```python
df['size'].unique()
# Quantitative Columns = total_bill, tip
```

```python
plt.subplot(1,2,1) # (row, col, which block)
sns.histplot(df["total_bill"], bins = 20, kde= True)
plt.subplot(1,2,2)
sns.histplot(df['tip'], bins = 20 , kde= True)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'scatter')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'kde')
```

```python
sns.jointplot(x = 'total_bill', y='tip' , data=df, kind= 'reg')
# sns.jointplot(x=df['total_bill'] , y=df['tip'] , kind='reg') 
# Both ways Works
```

```python
# sns.pairplot(df) #automatically make compare charts
sns.pairplot(df, hue='sex')
```

```python
sns.pairplot(df,hue='size',palette='rainbow')
```

```python
sns.rugplot(df['tip'])
```
