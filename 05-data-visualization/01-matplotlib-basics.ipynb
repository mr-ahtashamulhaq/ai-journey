**Matplotlib Basics**

```python
x = [1,2,3,4,5]
y = [10,50,30,45,70]
plt.plot(x,y)
plt.title("It is name of the Chart")
plt.xlabel("It is X val")
plt.ylabel("It is Y val")
```

```python
x = np.linspace(0,5,11)
x
y = x **3 
y
```

```python
plt.plot(x,y)
plt.title("Exam Marks")
plt.xlabel("exam")
plt.ylabel("score")
```

**Subplots in Matplotlib**

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(y,x)
plt.subplot(2,2,3)
plt.plot(x,x)
plt.subplot(2,2,4)
plt.plot(y,x**x)
```

**Object Oriented way**

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches
dpi is the dots-per-inch (pixel per inch).

```python
fig = plt.figure(figsize=(20,10),dpi = 50) #(height, width) of figure
axis1 = fig.add_axes([0.1,0.1,.8,.8]) # ([left, bottom, width, height]) Each value goes from 0 to 1.
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.1,.3,.3])
axis2.plot(x,y)
axis3 = fig.add_axes([.7,.7,.3,.3])
axis3.plot(x,y)
```
