```python
import seaborn as sns 
import matplotlib.pyplot as plt 
```

# Heat map

```python
flights = sns.load_dataset('flights')
flights
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
# We use heatmap to show correlation.
# To calculate correlation we take all neumeric values only and apply .corr() func

tipscorr  = tips[['total_bill','tip','size']] # neumeric vals in table
```

```python
tipscorr
```

```python
tipscorr.corr()
```

```python
sns.heatmap(tipscorr.corr())
```

```python
sns.heatmap(tipscorr.corr(), annot=True)
```

# Cluster map 

```python
sns.clustermap(tipscorr.corr())
```
