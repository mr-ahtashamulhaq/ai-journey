## Steps we do to clean data
1. Deleting Redundant columns
2. Renaming Columns
3. Dropping Duplicates
4. Remove the NaN values from Dataset
5. Check for some more transformation
---


### 1. Deleting Redundant columns

```python
data.columns
```

```python
# Delete Unnamed column
data.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
data.columns
```

### 2. Renaming Columns

```python
# Rename the deatils -> details
data.rename(columns={"Deatils" :"Details" },inplace=True)
```

```python
data.columns
```

### 3. Dropping Duplicates

```python
# It will shows us how many rows are duplicated
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.duplicated().sum()
```

### 4. Remove the NaN values from Dataset

```python
# It will show us any duplicate value we have
data.isna().sum()
```

```python
# So we don't have any NULL values But if we have we can remove by this
data.dropna(inplace=True)
```

### 5. Check for some more transformation

```python
data.reset_index(inplace=True)
```

```python
data.head()
```

```python
# delete old index columns
data.drop(columns="index", inplace=True)
```

```python
# In Sizes column there are nan value but it didn't appeared because these are string
data["Sizes"][27929]
```

```python
# SO rename them to NA VALUED
data["Sizes"] = data["Sizes"].replace("Nan" , "NA VALUED")
```

```python
# Now remove these rows
data = data[data["Sizes"] != "NA VALUED"]
```

```python
data
```

```python
# In Discount column there are also Nan values , meaning it doesn't have any discount
# SO rename them to No Discount
data["Discount"] = data["Discount"].replace("Nan" , "No Discount")
data
```

```python
# Size column also have Size:Error Size , so replce it with No Size
data["Sizes"] = data["Sizes"].replace("Size:Error Size" , "No Size")
data
```

```python
data["MRP"]
```

```python
# MRP also has Nan values so in it we can fill Mean of all values
# first convert MRP values to int
data["MRP"] = data["MRP"].replace("Nan", "00000")
data["MRP"] = data["MRP"].str[3:]

```

```python
data["MRP"] = data["MRP"].astype(int)
```

```python
replaceval = data["MRP"].mean()
data["MRP"] = data["MRP"].apply(lambda x : int(replaceval) if x == 0 else int(x))
```

```python
data["MRP"]
```

```python
# Save the File
data.to_csv("cleaned_data.csv")
```
