### 1. Deleting Redundant columns

```python
# Now we check if there is any redundant columns
data.columns
```

```python
df.head()
```

```python
# We can drop rating
df.drop(columns=["rating"], inplace=True)
```

### 2. Renaming Columns

```python
df.columns
```

```python
# we make a dictionary with -- orgName : newName
new_names = {}
for i in data.columns:
    new_names[i] = i.replace("_"," ").title()
print(new_names)
```

```python
# now replace old names with new ones
data.rename(columns=new_names, inplace=True)
data
```

```python
df = pd.DataFrame(data)
```

### 3. Dropping Duplicates

```python
# Check if there's any duplicate rows
data.duplicated().sum()
# It says 0 mean we don't have any duplicate rows
```

```python
# In this case w don't have any duplicates
# But if we do have we can remove those using

data.drop_duplicates(inplace=True)
```

### 4. Check for NULL Values


```python
data.isna().sum()
```

```python
# it shows we don't have any null values
# but if we have, we can remove those using

data.dropna(inplace=True)
```

### 5. Check for some more transformation


```python
data.columns
```

```python
# nunique tells us total unique values in a particular column
data["Show Id"].nunique()
```

```python
# Meaning we don't have duplicates in it
# But in Show Id values are string we have to make them integer
df["Show Id"].head()
```

```python
# Remove 's'
data["Show Id"] = data["Show Id"].apply(lambda x : x[1:])
```

```python
# convert into integer
data["Show Id"] = data["Show Id"].astype(int)
```

```python
data["Show Id"]
```

```python
data["Date Added"]
```

```python
# Date have '/' this symbol , we have to change it with '-'
data["Date Added"] = data["Date Added"].apply(lambda i : i.replace("/", "-"))
```
