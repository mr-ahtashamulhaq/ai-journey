### 1. Deleting Redundant Columns

```python
# here we make list of which cols we want and which we want to drop
column_keep = ['NAME', 'host id', 'host_identity_verified', 'host name',
       'neighbourhood group', 'neighbourhood', 'lat', 'long', 'country',
       'country code', 'instant_bookable', 'cancellation_policy', 'room type',
       'Construction year', 'price', 'service fee', 'minimum nights',
       'number of reviews', 'last review']
column_drop= ['reviews per month',
       'review rate number', 'calculated host listings count',
       'availability 365', 'house_rules', 'license','id']
```

```python
print(len(column_keep))
print(len(column_drop))
```

```python
data.drop(columns=column_drop, inplace=True)
```

```python
data.shape
```

```python
df = pd.DataFrame(data)
df
```

### 2. Renaming Columns

```python
data.columns
```

```python
# Preparing dictionary which -- OriginalName : NewName
mapping_dic = {}
for i in data.columns:
    tempName = "_".join(i.split(" ")).lower()
    mapping_dic[i] = " ".join(tempName.split("_")).title()
print(mapping_dic)
```

```python
data.rename(columns=mapping_dic, inplace=True)
data.columns
```

### 3. Dropping Duplicates

```python
# It tells for each row either it is duplicated or not
data.duplicated() 
```

```python
# More detail
print("First 5 Rows:")
print(data.duplicated().head())
print("\nLast 5 Rows:")
print(data.duplicated().tail())
print("\nTotal Duplicated Rows:")
print(data.duplicated().sum())
print("\nTotal True and Total False Rows:")
print(data.duplicated().value_counts())
```

```python
# Drop duplicate rows
data.drop_duplicates(inplace=True)
```

```python
print(data.duplicated().sum())
# it will show 0 , means all duplicates removed
```

### 4. Remove NaN Values

```python
data.isna().sum()
```

```python
# Last Review has a lot of NULL values, so we drop it
data.drop(columns=["Last Review"], inplace=True)
```

```python
data.isna().sum()
```

```python
# Now Drop all Nan Rows
data.dropna(inplace=True)
```

```python
data.isna().sum()
```

### 5. Cleaning individual columns

```python
data
```

```python
# Make 'Host Identity Verified' column values to upper case
data["Host Identity Verified"] = data["Host Identity Verified"].str.upper()
data["Host Identity Verified"]
```

```python
# In column 'Instant Bookable' instead of True/False make 1/0
data["Instant Bookable"] = data["Instant Bookable"].apply(lambda x : 1 if x == True else 0)
data["Instant Bookable"]
```

```python
# We don't have consistent number of indexes so we make index consistent
data.reset_index(inplace=True)
data
```

```python
# Price column has values as string , we will converti it into int()
data["Price"].head()
```

```python
data["Price"] = data["Price"].str.replace("$", "")
data["Price"] = data["Price"].str.replace(",", "")
data["Price"] = data["Price"].str.replace(" ", "")

data["Price"]
```

```python
data["Price"] = data["Price"].astype(int)
type(data["Price"][0])
```

```python
# Save this data as a new CSV Data
data.to_csv("cleaned_data.csv")
```
