# **Tensors?**

## 1. Introduction to Tensors
* **What is a Tensor?** A tensor is a fundamental data structure or a container for numbers. While occasionally used for characters, 99.99% of the time tensors store numerical values.
* **Why Learn About Tensors?** They are the core building blocks of Machine Learning and Deep Learning frameworks (PyTorch, TensorFlow, Scikit-Learn).
* **Generalization:** Tensor is a unified term across computer science and physics covering scalars, vectors, matrices, and multi-dimensional arrays.

## 2. Key Terminology
* **Rank / Axes / Dimensions:** The number of directions or dimensions a tensor has (e.g., a 2D matrix has a rank of 2).
* **Shape:** The number of elements along each axis (e.g., shape `(2, 3)` = 2 rows, 3 columns).
* **Size:** Total count of items inside the tensor, calculated by multiplying the dimensions in the shape.

---
## 3. Types of Tensors & Machine Learning Use Cases

### 3.1 0D Tensor (Scalar)
* **Definition:** A single numerical value (Rank: 0, Size: 1).

```python
import numpy as np

# 0D Tensor
tensor_0d = np.array(5)
print("Value:", tensor_0d)
print("Rank (ndim):", tensor_0d.ndim)
print("Shape:", tensor_0d.shape)
print("Size:", tensor_0d.size)
```

### 3.2 1D Tensor (Vector)
* **Definition:** An array or sequence of scalars.
* **ML Use Case:** Represents a single record or feature array (e.g., a student's `[CGPA, IQ, Test Score]`).

```python
# 1D Tensor (Single student record)
tensor_1d = np.array([3.8, 120, 85])
print("Rank (ndim):", tensor_1d.ndim)
print("Shape:", tensor_1d.shape)
print("Size:", tensor_1d.size)
```

### 3.3 2D Tensor (Matrix)
* **Definition:** A collection of 1D vectors arranged in rows and columns.
* **ML Use Case:** Tabular datasets (CSVs, SQL tables) where rows = samples and columns = features.

```python
# 2D Tensor (3 students, 3 features each)
tensor_2d = np.array([
    [3.8, 120, 85],
    [3.2, 110, 78],
    [3.9, 130, 92]
])
print("Rank (ndim):", tensor_2d.ndim)
print("Shape:", tensor_2d.shape)
print("Size:", tensor_2d.size)
```

### 3.4 3D Tensor
* **Definition:** A collection of 2D matrices stacked along a depth axis.
* **ML Use Cases:**
  1. **NLP:** Collection of text sequences (Sentences x Words x Embedding Vector).
  2. **Time Series:** Stock prices tracking (Companies x Time x Features).

```python
# 3D Tensor Example (Shape: 2 samples, 3 timesteps, 2 features)
tensor_3d = np.random.rand(2, 3, 2)
print("Rank (ndim):", tensor_3d.ndim)
print("Shape:", tensor_3d.shape)
print("Size:", tensor_3d.size)
```

### 3.5 4D Tensor
* **Definition:** A collection of 3D tensors.
* **ML Use Case:** Image Batches in Computer Vision (Batch Size x Height x Width x RGB Channels).

```python
# 4D Tensor Example (Batch of 32 images, 28x28 resolution, 3 RGB channels)
tensor_4d = np.random.rand(32, 28, 28, 3)
print("Rank (ndim):", tensor_4d.ndim)
print("Shape:", tensor_4d.shape)
print("Size:", tensor_4d.size)
```

### 3.6 5D Tensor
* **Definition:** A collection of 4D tensors.
* **ML Use Case:** Video Batches (Batch Size x Frames x Height x Width x RGB Channels).

```python
# 5D Tensor Example (4 videos, 10 frames each, 64x64 resolution, 3 RGB channels)
tensor_5d = np.random.rand(4, 10, 64, 64, 3)
print("Rank (ndim):", tensor_5d.ndim)
print("Shape:", tensor_5d.shape)
print("Size:", tensor_5d.size)
```
