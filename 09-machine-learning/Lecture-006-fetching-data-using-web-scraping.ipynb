```python
import pandas as pd 
import requests
from bs4 import BeautifulSoup
```

```python
headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/138.0.0.0 Safari/537.36"
    ),
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8",
    "Accept-Language": "en-US,en;q=0.9",
    "Accept-Encoding": "gzip, deflate, br",
    "Connection": "keep-alive",
    "Upgrade-Insecure-Requests": "1"
}
response = requests.get("https://pk.indeed.com/jobs?q=&l=lahore&radius=25&from=searchOnDesktopSerp&start=0", headers=headers)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
job_cards = soup.select('div.job_seen_beacon, td.resultContent')

data = []
for card in job_cards:
    
    title_el = card.select_one('h2.jobTitle span, a.jcs-JobTitle')
    company_el = card.select_one('[data-testid="company-name"], .companyName')
    salary_el = card.select_one('[data-testid="attribute_snippet-label"], .salary-snippet-container')

    
    title = title_el.text.strip() if title_el else None
    company = company_el.text.strip() if company_el else None
    salary = salary_el.text.strip() if salary_el else "Not Disclosed"

    if title:
        data.append({"Title": title, "Company": company, "Salary": salary})

df = pd.DataFrame(data)
```

```python
df = df.drop_duplicates(subset=['Title', 'Company'], keep='first').reset_index(drop=True)
df
```
