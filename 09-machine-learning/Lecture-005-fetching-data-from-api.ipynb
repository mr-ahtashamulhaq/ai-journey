```python
import pandas
import requests
```

```python
# in the last of url we have mentioned page=1 means we want to get page 1 data
url = "https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1"
response = requests.get(url)
```

**Till here we only got page 1 data from Api But there are 428 different pages so we have to run a loop and get data from all pages and append in one data Frame**

```python
df = pd.DataFrame()

for i in range(1,429):
    url = f"https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={i}"
    response = requests.get(url)
    temp_df = pd.DataFrame(response.json()['results'])[['id', 'title', 'overview', 'release_date', 'popularity']]
    df = pd.concat([df, temp_df], ignore_index=True)
    
```
